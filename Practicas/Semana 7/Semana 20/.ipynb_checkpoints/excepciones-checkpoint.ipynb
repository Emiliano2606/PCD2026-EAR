```python
#Convertidor de grados Kelvin (Â°K) a grados centrigrados (Â°C) y Farenheit (Â°F)
#Uso de try y excep
Valor_Kelvin = None
bandera = False
try:
    Valor_Kelvin = float (input("Ingresa el numero que quieres convertir: "))
    if (Valor_Kelvin < 0):
        print ("El numero debe ser mayor a 0")
    else:
        bandera = True
except ValueError:
    print("Error de conversion, asegurese que el valor sea un numero real mayor a cero")

G_Celcius = Valor_Kelvin - 273.15
G_Farenheit = (G_Celcius * 9/5) + 32
print (f"La temperatura en grados Celcius es: {G_Celcius:.3f}Â°C\n")
print (f"La temperatura en grados Farenheit es:{G_Farenheit:.3f} Â°F")


    
```

```python
#Lectura de un archivo
try:    
    archivo = open("Lectura.txt", 'r', encoding='utf-8') 
    contenido = archivo.read()
    
except FileNotFoundError:
    print(f"\nError: El archivo no fue encontrado.")

else:
    print (f"El archivo {"Lectura.txt"} contiene lo siguiente:\n")
    print (contenido)

finally:
    print("\nCerrando el archivo")
    if archivo is not None and not archivo.closed:
        archivo.close()
```
