**Ejemplos de lectura y escritura en archivos .txt**

```python
#Creación de un archivo .txt
archivo = open("Prueba.txt", "w")
```

```python
#Agregar una nueva linea de texto e imprimir su contenido
with open("Prueba.txt", "w") as f:
    f.write("Este es un ejemplo, el archivo original esta vacío\n:)")
```

```python
with open("Prueba.txt", "r") as f:
    contenido = f.read()
    print(contenido)
```

```python
#Sobreescribir una linea
with open("Prueba.txt", "w") as f:
    f.write("Vamos a sobreescribir lo que estaba en el texto")
```

```python
with open("Prueba.txt", "r") as f:
    contenido = f.read()
    print(contenido)
```

```python
#Ahora vamos a leerlo y escribirlo
with open("Prueba.txt", "r+") as f:
    f.seek(0, 2)  # mueve el puntero al final del archivo
    f.write("\nNueva línea al final.")
    f.seek(0)     # opcional: vuelve al inicio para leer todo
    contenido = f.read()

print(contenido)
```

**Ejemplos de lectura y escritura de archivos .csv**

```python
#Creación de un archivo .csv
import csv
with open('Base.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Animal", "Nombre", "Raza", "Peso (kg)", "Edad (años)"])
    writer.writerow(["Gato", "Luna", "Siamés", 4.5, 3])
    writer.writerow(["Gato", "Tigre", "Maine Coon", 7.0, 5])
    writer.writerow(["Perro", "Max", "Labrador", 32.5, 7])
    writer.writerow(["Perro", "Chispa", "Chihuahua", 2.1, 2])
    writer.writerow(["Pez", "Burbuja", "Pez Payaso", 0.1, 1])
    writer.writerow(["Pez", "Nemo", "Guppy", 0.05, 0.5])


    
```

```python
#Ahora vamos a imprimirlo
with open('Base.csv', mode='r', newline='', encoding='utf-8') as file:
    lector = csv.reader(file)
    for fila in lector:
        print(fila)
```

```python
#Vamos a agregar una nueva linea
with open('Base.csv', mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow (["Ave", "Dory", "Loro", 3.2, 2.1])
    
    
    
    
        
```

```python
#Ahora vamos a imprimirlo
with open('Base.csv', mode='r', newline='', encoding='utf-8') as file:
    lector = csv.reader(file)
    for fila in lector:
        print(fila)
```

```python
#Ahora vamos a copiar estos datos a un txt
with open('Base.csv', mode='r', encoding='utf-8') as FileCsv:
    with open('CopiaCsv.txt', mode='w', encoding='utf-8') as FileTxt:
        contenido = FileCsv.read()
        FileTxt.write(contenido)
    
    
```

```python
#Ahora vamos a eliminar el csv original y lo vamos a crear a partir del txt
import os
try:
    os.remove("Base.csv")
    print("Archivo eliminado exitosamente")
        
except FileNotFoundError:
    print(f"Error: El archivo '{Base.csv}' no fue encontrado.")

#Ahora vamos a crearlo 
```

```python
#Ahora vamos a crear el csv a partir del txt
with open('CopiaCsv.txt', mode='r', encoding='utf-8') as CopiaFiletxt:
    with open('Copia_Base.csv', mode='w', encoding='utf-8') as CopiaCsv:
        contenido = CopiaFiletxt.read()
        CopiaCsv.write(contenido)
        
```

```python
#Vamos a agregar una nueva nueva al txt
with open('CopiaCsv.txt', mode='a', encoding='utf-8') as file:
    file.write("Reptil,Policarpio,Iguana,7.8,4.1\n")
```

```python
with open('Copia_Base.csv', mode='r', newline='', encoding='utf-8') as file:
    lector = csv.reader(file)
    for fila in lector:
        print(fila)
```
