## Crea un notebook llamado poo_basica.ipynb que contenga al menos 2 clases propias, por ejemplo:

- Coche: atributos (marca, modelo, año), método arrancar().
- Libro: atributos (título, autor, páginas), método descripcion().
- Empleado: atributos (nombre, salario), método aumento(porcentaje).
- Cada clase debe incluir:

Un método __init__ (constructor).
Al menos un método adicional.
Ejemplos de instanciación y uso.

```python
class Libro:
    def __init__ (self,titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.pagina = paginas
    def descripcion (self):
        return f"El libro: {self.titulo}\nDel autor: {self.autor}\nTiene un total de: {self.pagina}"
    def Tiempo_lectura (self):  
        return f"\nEl libro {self.titulo} tiene un tiempo de lectura de {self.pagina * 2} minutos"

Libro1 = Libro("El principito", "Antoine de Saint-Exupéry", 111 )
print (Libro1.descripcion())
print (Libro1.Tiempo_lectura())
        
```

```python
class Avion:
    def __init__ (self, Aerolinea, Num_boleto, Hora_salida, Distancia=0):
        try:
            if (Num_boleto <0):
                raise ValueError("ERROR NUMERO DE BOLETO: El número de boleto no puede ser negativo")
                
            self.Aerolinea = Aerolinea
            self.Num_boleto = Num_boleto
            self.Hora_salida = Hora_salida
            self.Distancia = Distancia

        except ValueError as e:
            print(e)
            raise  
    def Tiempo_llegada (self, Distancia):
        return f"Su vuelo de la aerolinea {self.Aerolinea} parte a las {self.Hora_salida} horas, y tiene un tiempo estimado de {(self.Distancia / 860):.2f} horas"
   
    def Tipo_clase (self):
        if (self.Num_boleto < 100):
            return f"Su boleto con numero: {self.Num_boleto} \nPERTENECE A PRIMERA CLASE"
        elif (self.Num_boleto > 100 and self.Num_boleto <500):
            return f"Su boleto con numero: {self.Num_boleto} \nPERTENECE A CLASE MEDIA"
        else:
            return f"Su boleto con numero: {self.Num_boleto} \nPERTENECE A CLASE TURISTA"
        
        
    
Avion1 = Avion ("Aeromexico", 5, 12, 9093)
print (Avion1.Tiempo_llegada(9093))
print (Avion1.Tipo_clase ())
```
