```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejemplo 1 - Slicing b치sico 2D

```python
arr = np.arange(1, 25).reshape(4, 6)
print("Array original:\n", arr)
print("Filas 1 a 3 (indice 1:3), columnas 2 a 5: \n", arr[1:3, 2:5])
```

## ejemplo 2 - Slicing con step (paso)

```python
print("Cada segunda fila y cada segunda columna:\n", arr[::2, ::2])
print("Array invertido:\n", arr[::-1, ::-1])
```

## Ejemplo 3- Indexaci칩n con listas (fancy indexing)

```python
filas = [0, 2, 3]
columnas = [1, 4, 5]
print("Elementos especificos con listas de indices:\n", arr[filas, columnas]) #Nos permite acceder a elementos mediante los indices
print("Submatriz usando listas:\n", arr[[0, 1, 3]]) # solo estamos tomando las filas
```

## Ejemplo 4 - M치scara booleana simple

```python
mask = arr % 3 == 0
print("M치scara (m칰ltiplos de 3):\n", mask) # Devuelve el array con los booleanos de la mascara
print("Valores que son m칰ltiplos de 3:\n", arr[mask]) # Devuelve los valores que se encuentran en los indices de los True
print("Cantidad de m칰ltiplos de 3:\n", arr[mask].size) # Esto devuelve la cantidad de elementos que son multiplos de 3
```

## Ejemplo 5 - M치scara booleana combinada + modificaci칩n

```python
mask2 = ((arr > 10) & (arr < 20))  #esta m치scara dara true a los valores que se encuentran entre 11 y 19
print("Array de booleanos:\n", mask2)
print("Valores del 11 al 19: \n", arr[mask2])

# Modificacion de valores
arr_mod = arr.copy()# creamos una copia del arreglo original para demostrar que se puede alterar
print("Array copiado:\n", arr_mod)
arr_mod[mask2] = 999  #Reemplazamos los valores con 999
print("Array modificado (11-19 -> 999):\n", arr_mod)
```
