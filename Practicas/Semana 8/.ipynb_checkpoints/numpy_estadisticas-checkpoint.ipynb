# Funciones matematicas elemenetales (Ufuncs)

las ufuncs (universal functions) aplican operaciones elemento a elemento de forma vectorizada como si fuera un escalar.

- `np.add`, `np.subtract`, `np.multiply`, `np.divide`, `np.power`
- Todas aceptan *broadcasting* cuando las formas son compatibles.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

#uso de las funciones
print(a)
print(b)
print("add: ",np.add(a, b))
print("add:",a + b)
print("Substract:", np.subtract(b,a))
print("Multiply:", np.multiply(b,a))
print("Divide:", np.divide(b,a))
print("Divide:", np.divide(a, b))
print("power:", np.power(b, a))
```

```python
print(a.sum())
print(np.sum(a))
```

```python
m = np.arange(20).reshape(5, 4)
print(m)
```
