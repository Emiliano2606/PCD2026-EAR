# üîπ 1. Conjuntos de datos aleatorios + estad√≠sticas

```python
# Ejemplo 1
data1 = np.random.rand(50)
print(data1)
print("Media aritmetica:",np.mean(data1))
print("Mediana:", np.median(data1))
print("Desviaci√≥n estandar:", np.std(data1))
print("Valor minimo:", np.min(data1))
print("Valor maximo:", np.max(data1))
```

```python
# Ejemplo 2
data2 = np.random.randn(100)
print(data2)
print("Media aritmetica:",np.mean(data2))
print("Mediana:", np.median(data2))
print("Desviaci√≥n estandar:", np.std(data2))
print("Valor minimo:", np.min(data2))
print("Valor maximo:", np.max(data2))
```

```python
# Ejemplo 3
data3 = np.random.randint(0, 100, size=30)
print(data3)
print("Media aritmetica:",np.mean(data3))
print("Mediana:", np.median(data3))
print("Desviaci√≥n estandar:", np.std(data3))
print("Valor minimo:", np.min(data3))
print("Valor maximo:", np.max(data3))
```

```python
# Ejemplo 4 (distribuci√≥n uniforme en rango distinto
data4 = np.random.uniform(-10, 10, size=40)
print(data4)
print("Media aritmetica:",np.mean(data4))
print("Mediana:", np.median(data4))
print("Desviaci√≥n estandar:", np.std(data4))
print("Valor minimo:", np.min(data4))
print("Valor maximo:", np.max(data4))
```

```python
# Ejemplo 5 (normal con media y sigma diferente
data5 = np.random.normal(loc=20, scale=5, size=60)
print(data5)
print("Media aritmetica:",np.mean(data5))
print("Mediana:", np.median(data5))
print("Desviaci√≥n estandar:", np.std(data5))
print("Valor minimo:", np.min(data5))
print("Valor maximo:", np.max(data5))
```

# üîπ 2. Matriz 5x5 y an√°lisis por axis

```python
import numpy as np

np.random.seed(42)

# Matriz 5x5
M = np.random.randint(1, 20, (5, 5))
print("Matriz 5x5:\n", M)

# Sumas
print("Suma por filas (axis=1):", np.sum(M, axis=1))
print("Suma por columnas (axis=0):", np.sum(M, axis=0), "\n")

# Medias
print("Media por filas (axis=1):", np.mean(M, axis=1))
print("Media por columnas (axis=0):", np.mean(M, axis=0), "\n")

# M√≠nimos
print("M√≠nimo por filas (axis=1):", np.min(M, axis=1))
print("M√≠nimo por columnas (axis=0):", np.min(M, axis=0), "\n")

# M√°ximos
print("M√°ximo por filas (axis=1):", np.max(M, axis=1))
print("M√°ximo por columnas (axis=0):", np.max(M, axis=0))
```

# üé≤ 3. Simulaci√≥n de 1000 lanzamientos de dado

```python
lanzamientos = np.random.randint(1, 7, size = 1000)
valores, conteos = np.unique(lanzamientos, return_counts =True)
porcentajes = conteos / 1000 * 100

print("Resultados de 1000 lanzamientos:")
print("Caras:", valores)
print("Frecuencias:", conteos)
print("Porcentajes:", porcentajes)
    
```

# üîé Conclusi√≥n:

```python
# rand ‚Üí n√∫meros uniformes entre 0 y 1
a = np.random.rand(5)

# randn ‚Üí n√∫meros con distribuci√≥n normal media 0, std 1
b = np.random.randn(5)

# randint ‚Üí enteros aleatorios en un rango
c = np.random.randint(1, 10, size=5)

print(a)
print(b)
print(c)
```

- **`rand`**: √∫til cuando necesitas generar n√∫meros con **distribuci√≥n uniforme entre 0 y 1**. Ideal para simular proporciones, probabilidades o cualquier variable que deba ser uniformemente distribuida.

- **`randn`**: se usa cuando ocupas datos con **distribuci√≥n normal (media 0 y varianza 1)**. Muy √∫til en modelos estad√≠sticos, ruido gaussiano o cualquier proceso que siga una distribuci√≥n normal.

- **`randint`**: es la mejor opci√≥n cuando necesitas **valores enteros**, como simulaciones de dados, selecci√≥n de √≠ndices, generaci√≥n de IDs o cualquier proceso discreto.
