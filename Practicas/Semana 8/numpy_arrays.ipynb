## Ejemplos de arrays con funciones

```python
import numpy as np

# arrays a partir de lista.
lista1 = [10, 20, 30, 40, 50]
arr1 = np.array(lista1)
print("Arrays a parti de lista:")
print(arr1)


# arrays de ceros
arr_zeros = np.zeros(5)
print("Arrays de ceros:")
print(arr_zeros)


# array de unos (Matriz:2x3)
arr_unos = np.ones((2, 3))
print("Arrays de unos:")
print(arr_unos)

# Array con arange
arr_arange = np.arange(0, 31, 3)
print("Array con arange:")
print(arr_arange)

# Array con linspace
arr_lins = np.linspace(1, 10, 5)
print(" Array con linspace:")
print(arr_lins)

# Array con valores randoms
arr_random = np.random.rand(4,3)
print("Array con valores random:")
print(arr_random)

# Array identidad
arr_eye = np.eye(5)
print("Array identidad:")
print(arr_eye)
```

## ðŸ†š ComparaciÃ³n con listas.

```python
# Lista normal de python
lista = [10, 20, 30, 40, 50]
# array en numpy
x = np.array(lista)
# impresiÃ³n
print("Lista original: ",lista)
print("array numpy: ",x)

# MultiplicaciÃ³n: diferentes comportamientos
print("Lista *2: ", lista*2)
print("Array *2: ", x*2)

# suma de 10 a cada elemento
print("Lista + 10 â†’ Error? Prueba con list comprehension:")
print("print mediante for: ", [(x+10) for x in lista])
print("Array + 10 : ",  x +10)

# Velocidad (ejemplo pequeÃ±o)
import time

lista_grande = list(range(100_000_000))
array_grande = np.array(lista_grande)

# Multiplicar por 2 con lista
start = time.time()
lista_doble = [x * 2 for x in lista_grande]
end = time.time()
print(f"Tiempo con lista: {end - start:.4f} seg")

# Multiplicar por 2 con array
start = time.time()
array_doble = array_grande * 2
end = time.time()
print(f"Tiempo con array: {end - start:.4f} seg")
```

## 4. Uso de propiedades `shape` `size` `ndim` `dtype`

```python
# crear array de 2 dimensiones
matriz = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("Matriz: \n", matriz)
print("Shape (forma):\n", matriz.shape)
print("NÃºmero total de elementos:\n", matriz.size)
print("NÃºmero de dimensiones:", matriz.ndim)
print("Tipo de datos", matriz.dtype)

# Cambiar tipo de datos
matriz_float = matriz.astype(float)
print("Matriz:\n", matriz_float)
print("Tipos de datos: ", matriz_float.dtype)
```
