# Ejercicio 1 Generacion de matrix 2x2 medicante funcion

```python
#Crear una malla de coordenadas X, Y
x = np.array([1,2,3,4,5])
print (x.shape)

y= np.array ([20,22,24,26,28,30])
print (y.shape)
```

```python
#Ahora le vamos aplicar el newaxis
x = x[:, np.newaxis]
print (x.shape)

y = y[np.newaxis, :]
print (y.shape)
```

```python
#Hecho eso ahora si podemos operarlas, la funcion sera f(x,y) = x^2 + 5y
z = x**2 + 5*y
print (z)
```

# Estandarizacion por renglones

```python
X = np.random.randint(1, 250, size=(5, 5)).astype(float)
print (X)
#En una matriz 2x2 el renglo esta dadl por axis=1
media = X.mean(axis=1)
estandar = X.std(axis=1)
matrix_normalizada = (X - media)/estandar
print (matrix_normalizada)
```
