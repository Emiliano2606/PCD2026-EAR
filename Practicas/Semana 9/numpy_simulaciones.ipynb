```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(41352)
np.__version__
```

```python
#Se lanza 2 dados con 12 caras cada uno, se van a realizar 10,000 lanzamientos y se deben analizar los resultados de sus sumas

n=10000
dado1 = np.random.randint(1, 12, size=n)
dado2 = np.random.randint(1, 12, size=n)
suma = dado1 + dado2
#Los resultados posibles va de 2 hasta 24

valores = np.arange(2, 25)
conteo = np.bincount(suma, minlength=25)[2:]

plt.figure()
plt.bar(valores, conteo)
plt.title("Distribuci贸n de la suma de 2 dados (10,000 lanzamientos)")
plt.xlabel("Suma")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Analisis de estadisticas
#Medidas de distribucucion (media y mediana)
media = np.mean(suma)
mediana = np.median(suma)
print (media)
print (mediana)
```

```python
#Medidas de dispersion (varianza, desviacion estandar, rango)

varianza = np.var(suma)
desv = np.std(suma)
rango = np.max(suma) - np.min(suma)

print (varianza, desv, rango)
```
