## Ahora vamos a realizar el ejercicio de las relaciones

Experimento de correlaci贸n: genera tres variables (A, B, C) donde B dependa linealmente de A y C sea ruido; calcula y comenta la matriz de correlaci贸n 3x3.

```python
Matriz_correlacion =  np.random.randint(9,size=(3,3))
print(Matriz_correlacion)
print("El tamanio es:",Matriz_correlacion.shape)
print("Las dimensiones son:",Matriz_correlacion.ndim)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(18112025)

A = np.linspace(0, 50, 50) + np.random.randn(50) // 3
B = - A * 3.14151692
C = np.random.randn(50) + 13

data = np.column_stack((A, B, C))

print("Forma de la matriz de datos (Filas x Variables):", data.shape)

corr_matrix = np.corrcoef(data, rowvar=False)

print("\nMatriz de correlaci贸n 3x3:")
print(corr_matrix)

plt.figure(figsize=(6, 5))

plt.imshow(corr_matrix, cmap='plasma', interpolation='none')

plt.colorbar(label='Coeficiente de correlaci贸n')

plt.title("Matriz de correlaci贸n entre 3 variables (A, B, C)")

plt.xticks(range(3), ["A", "B", "C"])
plt.yticks(range(3), ["A", "B", "C"])

plt.show()
```

Esto nos dice que la relacion entre A y B es inversa, mientras que la de A con C y B con C no cercanas a cero, por lo que no tienen correlacion
