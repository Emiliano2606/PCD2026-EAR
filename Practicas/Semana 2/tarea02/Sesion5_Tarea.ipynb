# Diferencias entre Jupyter y Colab.
## 1. Jupyter / JupyterLab (Local)

-Ubicación y Ejecución: Es una aplicación que se instala y ejecuta directamente en tu propia computadora (ya sea Windows, Linux o Mac).

-Recursos: Utiliza la CPU, la RAM y el espacio en disco de tu máquina. Si tu computadora es lenta, el análisis de datos será lento.

-Entorno: Trabaja con tu propio entorno Python aislado (como tu .venv o un entorno Conda). Tienes control total sobre las versiones y las dependencias.

-Archivos: Accede a los datos y archivos directamente desde tu disco duro local.

Costo: Es completamente gratis, pero el rendimiento depende de tu hardware.

## 2. Google Colaboratory (Colab) (Nube)

-Ubicación y Ejecución: Es un servicio en la nube proporcionado por Google. Se ejecuta en sus servidores remotos.

-Recursos: Te permite usar la potencia de servidores de Google, incluyendo acceso gratuito a GPUs (unidades de procesamiento gráfico) y TPUs (esencial para machine learning).

-Entorno: Utiliza un entorno Linux preconfigurado por Google. Puedes instalar librerías temporalmente, pero el entorno se reinicia después de un tiempo de inactividad.

-Archivos: Se conecta y accede a archivos guardados en tu Google Drive o en un almacenamiento temporal.

-Costo: Es gratis para uso básico y ofrece opciones de pago para recursos más potentes o sesiones más largas.

# Un ejemplo en Python con Pandas cargando un dataset pequeño (CSV).

```python
import pandas as pd

# Define el nombre de tu archivo
# ¡Asegúrate de que este nombre (incluyendo mayúsculas) coincida exactamente!
nombre_archivo = 'datos_prueba.csv' 

# 1. Cargar el CSV usando pd.read_csv()
# Los datos del archivo se convierten en un DataFrame llamado df_datos.
df_datos = pd.read_csv(nombre_archivo)

# 2. Mostrar la verificación: Usamos .head() para confirmar la carga.
print("DataFrame 'datos_prueba.csv' cargado exitosamente:")
print(df_datos.head())

# 3. (Opcional) Mostrar la estructura
print("\nEstructura del DataFrame:")
df_datos.info()
```
