```python
import pandas as pd
import numpy as np
import random
```

```python
#Serie 1 Veterinaria
clinica = pd.Series (["Bigotes", "Firulais", "Doradito"], index=["Gato","Perro", "Pez"])
clinica
```

```python
#Serie 2 Cuentas de bancos
Cuentas = pd.Series ([25039453, 192291, 458206, 132742,0], index=["Cuenta A","Cuenta B", "Cuenta C", "Cuenta D", "Cuenta E"])
Cuentas
```

```python
#Series 3 las serie con listas separadas
lista_1 = np.random.randint(0,100, size=10)
lista_2 = ["Valor 1","Valor 2","Valor 3","Valor 4","Valor 5","Valor 6","Valor 7","Valor 8","Valor 9","Valor 10"]
relacion = pd.Series (lista_1, index = lista_2)
relacion
```

```python
#Series 4 Crear serie hibrida
serie_hibrida = pd.Series (["Hola", "Mundo", "Tardis", "Doctor", "Animalogo", "Perro banana", "telefono", "garrafonsito", "contacto", "Natalia Quintana"]
                           , index = lista_2)
serie_hibrida
```

```python
#Serie 5 con listas comprehesion
lista_compresion = [i for i in range(0,10)]
lista_aleatoria = np.random.randn (10)

serie_relacion = pd.Series (lista_aleatoria, index = lista_compresion)
serie_relacion
```
