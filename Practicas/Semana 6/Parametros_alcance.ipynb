# 1Ô∏è‚É£ Funci√≥n: potencia
def potencia(base, exponente=2):
    """
    Calcula la potencia de un n√∫mero.
    
    Par√°metros:
    base (float o int): n√∫mero base.
    exponente (int, opcional): exponente al que se elevar√° la base. Por defecto es 2.
    
    Retorna:
    float: resultado de base ** exponente.

````python

def potencia(base, exponente=2):
    x=base**exponente
    return x

print(potencia(3))        # 9
print(potencia(2, 3))     # 8
````

def precio_con_iva(precio, iva=0.16):
    """
    Calcula el precio total con IVA incluido.
    
    Par√°metros:
    precio (float): precio base del producto.
    iva (float, opcional): porcentaje de IVA (por defecto 0.16 -> 16%).
    
    Retorna:
    float: precio final con IVA.
    
    Ejemplo:
    >>> precio_con_iva(100)
    116.0
    >>> precio_con_iva(100, 0.08)
    108.0
    """
    return precio * (1 + iva)



````python
def precio_con_iva(precio,iva=0.16):
    x=precio+(iva*100)
    return x
    
print(precio_con_iva(100))        # 116.0
print(precio_con_iva(100, 0.08))  # 108.0
````

# 3Ô∏è‚É£ Funci√≥n: formatear_nombre
def formatear_nombre(nombre, apellido, mayus=False):
    """
    Devuelve el nombre completo formateado.
    
    Par√°metros:
    nombre (str): nombre de la persona.
    apellido (str): apellido de la persona.
    mayus (bool, opcional): si es True, convierte todo a may√∫sculas. Por defecto es False.
    
    Retorna:
    str: nombre completo formateado.
    
    Ejemplo:
    >>> formatear_nombre("juan", "perez")
    'Juan Perez'
    >>> formatear_nombre("juan", "perez", True)
    'JUAN PEREZ'
    """

````python
def formatear_nombre(nombre, apellido, mayus=False):
    x = nombre.strip()
    y = apellido.strip()

    if mayus==True:
        x = nombre.upper()
        y = apellido.upper()
    else:
        x = nombre.capitalize()
        y = apellido.capitalize()

    return x,y

print(formatear_nombre("emiliano", "ortiz"))         # Juan Perez
print(formatear_nombre("Tete", "lopez", True))  # MARIA LOPEZ
````

### üß† Bonus: ¬øPor qu√© evitar valores mutables como valores por defecto?

En Python, los **valores por defecto** de los par√°metros se **eval√∫an una sola vez**, en el momento en que se **define la funci√≥n**, no cada vez que se llama.  
Por eso, si usamos un objeto **mutable** (como una lista `[]`, un diccionario `{}`, o un conjunto `set()`), **todas las llamadas posteriores** compartir√°n el **mismo objeto en memoria**, lo que puede generar comportamientos inesperados.

#### üîç Ejemplo:
```python
def agregar_elemento(elemento, lista=[]):
    lista.append(elemento)
    return lista

print(agregar_elemento(1))  # [1]
print(agregar_elemento(2))  # [1, 2]  ‚Üê ¬°La lista se acumul√≥!
