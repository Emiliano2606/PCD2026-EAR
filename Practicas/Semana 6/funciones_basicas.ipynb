## 1) Cuadrado de un numero
Fórmula: numero*numero

```python

def cuadrado(n):
    return(f"El cuadrado es {n*n}")
    
print(cuadrado(4))
```

## 2) Invertir texto
Fórmula: Hacer un slicing

```python
def invertir(texto):
     texto=texto[::-1]
     return(f"El texto invertido es {texto}")

print(invertir("Hola"))
```

## 3) Contar vocales de un texto
Fórmula: Ver cuantas vocales hay en la palabra

```python
def invertir(texto):
    texto=texto.strip()
    vocales=["a","e","i","o","u"]
    total=0
    for vocal in texto:
        if vocal in vocales:
            total=total+1
    return(f"total de vocales {total}")

print(invertir("Hola"))
```

## 4) Ordenar una lista de numeros
Fórmula: Usar sorted que devuelve la lista ordenada

```python
def ordenar(list_numeros):
    x=sorted(list_numeros)
    return(f"Los numeros ordenados son{x}")
           
print(ordenar([5,2,3,2,56,4,651]))
```

## 5) Devolver numero maximo de una lista
Fórmula: verificar el numero maximo

```python
def maxnum(lista_numeros):
    maximo=lista_numeros[0]
    for numero in lista_numeros:
        if numero>maximo:
            maximo=numero
    return maximo
    
print(maxnum([1,2,3,4,5,6]))
print(maxnum([1,2,3,4,999,6]))
```

## 6) Funcion propia
Fórmula: dependiendo el nombre da el acceso a la puerta

```python
def acceso(nombre):
    permitidos = ["juan", "marco", "alexa"]
    nombre = nombre.strip().lower()
    
    if nombre in permitidos:
        print(f"Acceso permitido {nombre}")
    elif nombre == "":
        print("No ingresaste ningún nombre")
    else:
        print("Ingresa otro nombre por favor")

acceso("juan")
acceso("pedro")
```
