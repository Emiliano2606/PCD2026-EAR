```python
import pandas as pd
# DataFrame base con información de ventas
ventas = pd.DataFrame({
    "region": ["Norte", "Norte", "Sur", "Sur", "Este", "Este", "Oeste", "Oeste"],
    "mes": ["Enero", "Febrero", "Enero", "Febrero", "Enero", "Febrero", "Enero", "Febrero"],
    "producto": ["Laptop", "Laptop", "Mouse", "Laptop", "Mouse", "Mouse", "Laptop", "Mouse"],
    "categoria": ["Tecnología", "Tecnología", "Accesorios", "Tecnología",
                  "Accesorios", "Accesorios", "Tecnología", "Accesorios"],
    "ventas": [20000, 25000, 3000, 18000, 2500, 2800, 22000, 3200]
})
```

```python
ventas
```

```python
# Ventas totales por región y mes
pivot_region_mes = pd.pivot_table(
    ventas,
    values="ventas",
    index="region",
    columns="mes",
    aggfunc="sum",
    fill_value=0,
    margins=True
)

pivot_region_mes
```

```python
# Promedio de ventas por producto y región
pivot_producto_region = pd.pivot_table(
    ventas,
    values="ventas",
    index="producto",
    columns="region",
    aggfunc="mean",
    fill_value=0
)

pivot_producto_region
```

```python
# Múltiples métricas de ventas por categoría y mes
pivot_categoria_mes = pd.pivot_table(
    ventas,
    values="ventas",
    index="categoria",
    columns="mes",
    aggfunc=["sum", "mean", "max"],
    fill_value=0
)

pivot_categoria_mes
```
