```python
import pandas as pd
# DataFrame de estudiantes
estudiantes = pd.DataFrame({
    "id_estudiante": [1, 2, 3, 4],
    "nombre": ["Ana", "Luis", "María", "Carlos"]
})

# DataFrame de calificaciones
calificaciones = pd.DataFrame({
    "id_estudiante": [1, 2, 5],
    "calificacion": [90, 85, 88]
})


print(estudiantes)

```

```python
print(calificaciones)
```

```python
inner_merge = pd.merge(estudiantes, calificaciones, on="id_estudiante", how="inner")
inner_merge
```

```python
left_merge = pd.merge(estudiantes, calificaciones, on="id_estudiante", how="left")
left_merge
```

```python
# RIGHT JOIN: todas las calificaciones aunque no exista el estudiante
right_merge = pd.merge(estudiantes, calificaciones, on="id_estudiante", how="right")
right_merge
```

```python
# OUTER JOIN: unión completa
outer_merge = pd.merge(estudiantes, calificaciones, on="id_estudiante", how="outer")
outer_merge
```

```python
ventas_enero = pd.DataFrame({
    "producto": ["Laptop", "Mouse"],
    "ventas": [10, 50]
})

ventas_febrero = pd.DataFrame({
    "producto": ["Laptop", "Teclado"],
    "ventas": [8, 30]
})
```

```python
ventas_enero
```

```python
ventas_febrero
```

```python
ventas_total = pd.concat([ventas_enero, ventas_febrero], axis=0)
ventas_total
```

```python
datos_producto = pd.DataFrame({
    "producto": ["Laptop", "Mouse", "Teclado"]
})

datos_precio = pd.DataFrame({
    "precio": [20000, 300, 500]
})
```

```python
datos_producto
```

```python
productos_completos = pd.concat([datos_producto, datos_precio], axis=1)
productos_completos
```

```python
clientes = pd.DataFrame({
    "nombre": ["Juan", "Laura", "Pedro"]
}, index=[1, 2, 3])

compras = pd.DataFrame({
    "total_compras": [500, 700]
}, index=[1, 3])

clientes.join(compras)
```

```python
empleados = pd.DataFrame({
    "id_empleado": [1, 2, 3],
    "nombre": ["Sofía", "Miguel", "Andrea"]
})

departamentos = pd.DataFrame({
    "id_empleado": [1, 3],
    "departamento": ["IT", "Recursos Humanos"]
})
```

```python
empleados
```

```python
departamentos
```

```python
# Convertir columna a índice y hacer join
empleados.set_index("id_empleado").join(
    departamentos.set_index("id_empleado")
)
```
