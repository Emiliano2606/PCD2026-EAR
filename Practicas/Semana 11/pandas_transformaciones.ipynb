# Importamos las librerias

```python
import pandas as pd
import numpy as np
```

# Dataset de prueba

```python
data = {
    "Nombre": ["Ana", "Luis", "María", "Carlos", "Ana"],
    "Edad": [23, 30, 27, 40, 23],
    "Ciudad": ["CDMX", "GDL", "Monterrey", "CDMX", "CDMX"],
    "Salario": [15000, 18000, 17000, 22000, 15000]
}

df = pd.DataFrame(data)
df
```

## Ejemplo: Creación de nuevas columnas

```python
# Salario anual
df["Salario_anual"] = df["Salario"] * 12
df
```

```python
# Clasificar empleados por edad:
df["Categoria_Edad"] = np.where(df["Edad"] >= 30, "Adulto", "Joven")
df
```

 ## Ejemplo: Transformaciones con apply() y lambda

```python
# Convertir nombres a mayúsculas
df["Nombre_mayus"] = df["Nombre"].apply(lambda x: x.upper())
df
```

# Crear una función personalizada
def clasificar_salario(sal):
    if sal < 16000:
        return "Bajo"
    elif sal < 20000:
        return "Medio"
    else:
        return "Alto"

df["Nivel_salario"] = df["Salario"].apply(clasificar_salario)
df

## Ejemplo: Reemplazo de valores con map() o replace()

```python
# Reemplazo con map() // Convertir ciudades a completo
ciudades_map = {
    "CDMX": "Ciudad de Mexico",
    "GDL": "Guadalajara",
    "Monterrey": "MTY"
}

df["Ciudad_invertido"] = df["Ciudad"].map(ciudades_map)
df
```

```python
# Reemplazo con replace() // Cambiar etiquetas de categoría
df["Categoria_Edad"] = df["Categoria_Edad"].replace({"Joven": "J", "Adulto": "A"})
df
```
