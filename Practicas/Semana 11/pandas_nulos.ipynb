## DataFrame de pruebas

```python
data = {
    "Cliente": ["Ana", "Luis", "Ana", "Carlos", "Luis", "Marta", None],
    "Edad": [25, None, 25, 40, 33, None, 22],
    "Ciudad": ["CDMX", "Monterrey", "CDMX", None, "GDL", "GDL", "CDMX"],
    "Compras_mensuales": [5, 3, 5, None, 2, 4, None]
}

df = pd.DataFrame(data)
df
```

## Identificar valores nulos

```python
df.isnull().sum()
```

```python
df.isnull()
```

## ImputaciÃ³n de valores nulos

### Imputacion con media

```python
media_compras = df["Compras_mensuales"].mean()
df["Compras_mensuales_media"] = df["Compras_mensuales"].fillna(media_compras)
df
```

### Imputacion con mediana

```python
#Aplicamos sobre columna numerica: edad
mediana_edad = df["Edad"].median()
df["Edad_mediana"] = df["Edad"].fillna(mediana_edad)
df
```

### ImputaciÃ³n con valor fijo

```python
df["Ciudad_fija"] = df["Ciudad"].fillna("Desconocida")
df
```
